```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
from datasets import load_dataset
docs = load_dataset(f"Cohere/wikipedia-22-12-simple-embeddings", split="train")
```

```python
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess

corpus = [simple_preprocess(doc) for doc in docs['text']]
```

```python
import random
random.shuffle(corpus)
corpus2 = corpus[0:(len(corpus)//2)]
random.shuffle(corpus)
corpus3 = corpus[0:(len(corpus)//3)]
random.shuffle(corpus)
corpus4 = corpus[0:(len(corpus)//4)]
```

```python
model2 = Word2Vec(vector_size=384, min_count=1)
model3 = Word2Vec(vector_size=384, min_count=1)
model4 = Word2Vec(vector_size=384, min_count=1)
```

```python
model2.build_vocab(corpus2)
model3.build_vocab(corpus3)
model4.build_vocab(corpus4)
```

```python
import shutil
from IPython.display import FileLink
```

```python
model2.train(corpus2, total_examples=model2.corpus_count, epochs=model2.epochs)
os.mkdir('model2')
model2.save("model2/semantic-search_word2vec2.model")
shutil.make_archive("model2",'zip', "/kaggle/working/model2")
os.chdir(r'/kaggle/working')
FileLink(r'model2.zip')
```

```python
model3.train(corpus3, total_examples=model3.corpus_count, epochs=model3.epochs)
os.mkdir('model3')
model3.save("model3/semantic-search_word2vec3.model")
shutil.make_archive("model3",'zip', "/kaggle/working/model3")
os.chdir(r'/kaggle/working')
FileLink(r'model3.zip')
```

```python
model4.train(corpus4, total_examples=model4.corpus_count, epochs=model4.epochs)
os.mkdir('model4')
model4.save("model4/semantic-search_word2vec4.model")
shutil.make_archive("model4",'zip', "/kaggle/working/model4")
os.chdir(r'/kaggle/working')
FileLink(r'model4.zip')
```
